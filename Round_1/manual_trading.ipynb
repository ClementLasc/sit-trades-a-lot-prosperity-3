```python
import numpy as np 
import itertools
import math
```

# Rate Matrix

```python
rates = np.array([[1, 1.98, 0.64, 1.34],
                  [0.48, 1, 0.31, 0.7],
                  [1.49, 3.1, 1, 1.95],
                  [0.72,1.45, 0.52, 1]
                 ])
products = {0:'Seashells', 1:'Pizzas', 2:'Sillicon Nuggets', 3:'Snowballs'}
```

# Solution

We re gonna find the solution via bruteforce => the state space is 4+4^2+4^3+4^4+1 = 341

```python
def amount(seq):
    """Compute the final amount after a sequence of trades, starting with 1 SeaShell.

    Parameters
    ----------
    seq : list of int
        List of intermediate products traded.
    
    Returns
    -------
    float
        Payoff.
    """
    if not seq:
        return 1
    prod = rates[0, seq[0]] * rates[seq[-1], 0] # Set the first amount (1 Seashell)
    L = len(seq)
    for i in range(L-1):
        prod *= rates[seq[i], seq[i+1]]  # Compute the amount using the sequence of product / Rate Matrix
    return prod
```

```python
def maximize(L):
    """Among sequences of L intermediate products, compute the ones with greatest final amount.

    Parameters
    ----------
    L : int
        Number of intermediate products.
    
    Returns
    -------
    argmax : list of tuple
        Optimal sequences of intermediate trades.
    max_val : float
        Maximal final amount for L intermediate products.
    """
    seqs = itertools.product(*[range(0, 4) for _ in range(L)]) # Compute all the possible sequences
    max_val = float('-inf')
    argmax = []
    for seq in seqs: # For all sequences, compute the sequence with the best final amount
        p = amount(seq)
        if math.isclose(p, max_val):
            argmax.append(seq)
        elif p > max_val:
            max_val = p
            argmax = [seq]
    return (argmax, max_val)
```

```python
for L in range(0,5):
    print(maximize(L))
```

```python
argmax, val_max = maximize(4)
print("Possible Max profit:",(val_max-1)*100)
print("Optimal sequences of trades:")
for seq in argmax:
    res = ' -> '.join([products[0]] + [products[i] for i in seq] + [products[0]])
    print(res)
```
