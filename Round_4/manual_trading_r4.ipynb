```python
import matplotlib.pyplot as plt
import numpy as np

# Données extraites du tableau
data = [
    {'multiplier': 10, 'inhabitants': 1, 'players_percent': 0.998},
    {'multiplier': 80, 'inhabitants': 6, 'players_percent': 18.178},
    {'multiplier': 37, 'inhabitants': 3, 'players_percent': 5.118},
    {'multiplier': 17, 'inhabitants': 1, 'players_percent': 7.539},
    {'multiplier': 90, 'inhabitants': 10, 'players_percent': 11.807},
    {'multiplier': 31, 'inhabitants': 2, 'players_percent': 6.987},
    {'multiplier': 50, 'inhabitants': 4, 'players_percent': 8.516},
    {'multiplier': 20, 'inhabitants': 2, 'players_percent': 1.614},
    {'multiplier': 73, 'inhabitants': 4, 'players_percent': 24.060},
    {'multiplier': 89, 'inhabitants': 8, 'players_percent': 15.184}
]

# Calcul de la métrique et de la fréquence pour chaque entrée
for entry in data:
   
    entry['frequency'] = 100 * (entry['players_percent'] / 100)
    
    entry['ratio'] = entry['multiplier'] / (entry['inhabitants']+entry['frequency'] )

# Tri des données par fréquence pour l'affichage
data_sorted = sorted(data, key=lambda x: x['frequency'])

# Extraction des valeurs pour le graphique
frequencies = [entry['frequency'] for entry in data_sorted]
ratios = [entry['ratio'] for entry in data_sorted]
multipliers = [entry['multiplier'] for entry in data_sorted]

# Création du graphique
plt.figure(figsize=(12, 6))
plt.scatter(frequencies, ratios, color='blue', s=80, alpha=0.7)

# Ajout d'annotations pour chaque point (le multiplicateur)
for i, txt in enumerate(multipliers):
    plt.annotate(f"x{txt}", (frequencies[i], ratios[i]), 
                 xytext=(5, 5), textcoords='offset points')

# Ajout d'une courbe de tendance
z = np.polyfit(frequencies, ratios, 1)
p = np.poly1d(z)
plt.plot(np.linspace(min(frequencies), max(frequencies), 100), 
         p(np.linspace(min(frequencies), max(frequencies), 100)), 
         "--", color="red", alpha=0.7)

# Ajout des étiquettes et du titre
plt.xlabel('Frequency')
plt.ylabel('Ratio with choice')
plt.title('P&l distrib')
plt.grid(True, alpha=0.3)
plt.tight_layout()

# Affichage du graphique
plt.show()
```
